## **EXPLORATORY DATA ANALYSIS IN PYTHON**

# Importing the required libraries for EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading the data into the data frame

```python
df=pd.read_csv("/content/data.csv")
df.head()
```

```python
df.tail(15)
```

# Checking the types of data

```python
df.dtypes
```

# Dropping irrelevant columns

```python
pd.unique(df['Engine Fuel Type'])
```

```python
df = df.drop(['Engine Fuel Type'], axis=1)
df.head(5)
```

# Renaming the columns

```python
df=df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

# Dropping the duplicate rows

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

Now let us remove the duplicate data because it's ok to remove them.

```python
df.count()
```

```python
df=df.drop_duplicates()
df.head()
```

```python
df.count()
```

# Dropping the missing or null values

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

# Detecting Outliers

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Price'])
```

# Plot different features against one another

# Histogram

```python
df['Make'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(5,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Year'])
```

# Scatterplot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['HP'], y=df['Price'])
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['HP'],y=df['Price'])
```

# **Heat Maps**

```python
plt.figure(figsize=(10,5))
c= df.corr(numeric_only=True)
sns.heatmap(c,annot=True)
```
